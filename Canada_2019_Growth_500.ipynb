```python
#import selenium
from selenium import webdriver

#import urlib3
from bs4 import BeautifulSoup as bs

import pandas as pd
import numpy as np
import time
```

```python
driver_path = 'C:\Windows\chromedriver.exe'
driver = webdriver.Chrome(executable_path=driver_path)
print("Current session: {}".format(driver.session_id))
driver.get('https://www.canadianbusiness.com/lists-and-rankings/growth-500/2019-ranking-g500/')
time.sleep(5)

soup = bs(driver.page_source,'html.parser')
```

```python
table = soup.table
row = []
links = []
table_rows = table.find_all('tr')
columns = ['Rank','Company','5-year growth (%)','2018 revenue ($mil)','Industry','City','Province','Employees, 2018 (full-time equivalent)']
for tr in table_rows:
    td = tr.find_all('td')
    row.append([i.text for i in td])

df_company = pd.DataFrame(data = row, columns=columns)
df_company
```

```python
#export data in excel
df_company.to_excel('2019_Growth_500.xlsx',sheet_name='companies')
```
